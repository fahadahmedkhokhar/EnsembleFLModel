```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
models = ['AlexNet', 'ResNet18', 'ResNet50']
node_labels = ['Node 1', 'Node 2', 'Node 3', 'Node 4']
accuracies = np.array([
    [91.10, 97.80, 96.50],
    [86.00, 96.50, 96.40],
    [92.60, 97.70, 97.70],
    [87.00, 97.00, 97.50]
])

# Transpose the accuracy matrix
accuracies = accuracies.T

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.2
index = np.arange(len(node_labels))

for i, model in enumerate(models):
    ax.bar([x + i * bar_width for x in index], accuracies[i], bar_width, label=model)

ax.set_xlabel('Nodes')
ax.set_ylabel('Accuracy (%)')
ax.set_title('Accuracy of Models Across Nodes')
ax.set_xticks([idx + 1.5 * bar_width for idx in index])
ax.set_xticklabels(node_labels)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
models = ['AlexNet', 'ResNet18', 'ResNet50']
node_labels = ['Node 1', 'Node 2', 'Node 3', 'Node 4']
data = np.array([
    [91.10, 88.00, 92.00, 97.80, 88.00, 96.00, 96.50, 86.0, 97.10],
    [86.00, 93.00, 86.00, 96.50, 86.00, 97.00, 96.40, 87.00, 95.00],
    [92.60, 91.00, 93.00, 97.70, 98.00, 87.00, 97.70, 89.00, 96.00],
    [87.00, 76.00, 86.10, 97.00, 83.00, 97.00, 97.50, 90.10, 97.00]
])

# Reshape the data for plotting
data = data.reshape((len(node_labels), len(models), 3))

# Plotting
fig, axs = plt.subplots(len(node_labels), 1, figsize=(10, 14))

metrics = ['Accuracy', 'F-score', 'Precision']
colors = ['blue', 'green', 'red']

for i, node in enumerate(node_labels):
    for j, metric in enumerate(metrics):
        axs[i].bar(models, data[i, :, j], color=colors[j], label=metric)
        for idx, value in enumerate(data[i, :, j]):
            axs[i].text(idx, value, f'{value:.2f}', ha='center', va='bottom', fontsize=8, color='black')
    axs[i].set_ylabel(node)
    axs[i].legend()

plt.tight_layout()
plt.show()
```
